```python
import requests
```

```python
search_url = "https://search.realself.com/questions-api?offset={offset}&limit={limit}&sortBy=mostpopular&sortDirection=asc&term=&locationId=17550&procedureId=73088"
```

```python
search_url.format(offset=10, limit=20)
```

```python
res = requests.get(search_url.format(offset=1, limit=10))
```

```python
res
```

```python
len(res.json()['questions'])
```

```python
questions = []
for i in range(0, 2716, 10):
    res = requests.get(search_url.format(offset=i, limit=10))
    response_body = res.json()
    print(response_body['questions'])
    questions = questions + response_body['questions']
```

```python
questions
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(questions)
```
